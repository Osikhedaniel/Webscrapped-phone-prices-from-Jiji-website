```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from bs4 import BeautifulSoup
import pandas as pd
import time
```

```python

driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))


url = 'https://jiji.ng/mobile-phones?page='


phone_data = []
```

```python
for page_num in range(1, 7):  
    print(f"Scraping page {page_num}...")

   
    driver.get(url + str(page_num))

    
    time.sleep(5)

    
    page_source = driver.page_source

    
    soup = BeautifulSoup(page_source, 'html.parser')

    
    listings = soup.find_all('div', class_='b-list-advert__gallery__item')

    
    for listing in listings:
       
        name_tag = listing.find('div', class_='b-advert-title-inner')
        name = name_tag.text.strip() if name_tag else 'No Name'
        
       
        if name == 'No Name':
            name_tag_alt = listing.find('div', class_='b-list-advert-base__data__title')
            name = name_tag_alt.text.strip() if name_tag_alt else 'No Name'

        
        price_tag = listing.find('div', class_='qa-advert-price')
        price = price_tag.text.strip() if price_tag else 'No Price'

        
        link_tag = listing.find('a', href=True)
        link = 'https://jiji.ng' + link_tag['href'] if link_tag else 'No Link'

        
        location_tag = listing.find('span', class_='b-list-advert__region__text')
        location = location_tag.text.strip() if location_tag else 'No Location'

        
        phone_data.append({
            'Name': name,
            'Price': price,
            'Link': link,
            'Location': location
        })
```

```python
df = pd.DataFrame(phone_data)
```
